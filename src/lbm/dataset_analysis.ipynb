```python
import json
import pandas as pd
import numpy as np
from pathlib import Path
from typing import List, Dict
from pathlib import Path
import re
```

```python
# checking wp1_prompts_prepared.json

json_path = Path("../../data/interim/wp1_prompts_prepared.json")

with open(json_path, "r", encoding="utf-8") as f:
    data = json.load(f)

df = pd.DataFrame(data)
print("Loaded", len(df), "rows.")
df.head(2)
```

```python
# checking unique values for each specific fields

def show_unique_values(df):
    selected_columns = ['condition', 'model_name', 'refusal_flag', 'has_identity', 'regard_label', 'variant','attack_category', 'technique', 'wp1_test_result' ] 

    for col in selected_columns:
        print(f"Unique values in {col}:")
        print(df[col].unique())
        print()
```

```python
show_unique_values(df)
```
